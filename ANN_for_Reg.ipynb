# **ANN Regression model to predict the electrical energy output of a Combined Cycle Power Plant**

# 1) Import Libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

# 2) Data preprocessing

```python
#Importing dataset
dataset=pd.read_excel("Folds5x2_pp.xlsx") #Folder is in excel formate, so used read_excel
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
#Spiltting dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

## No feature Scaling as here Regression is there

# 3) Building ANN

```python
# Initalising ANN var
ann=tf.keras.models.Sequential()
```

```python
#Adding input layer and hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
#Adding second hidden Layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
#Adding output layer
ann.add(tf.keras.layers.Dense(units=1))
```

```python
#Compiling the ann
ann.compile(optimizer='adam',loss='mean_squared_error')
```

# 4) Training model


```python
ann.fit(X_train,Y_train,batch_size=32,epochs=100)
```

# 5) Prediction

```python
Y_pred=ann.predict(X_test)
```

```python
np.set_printoptions(precision=2) #For setting value of prediction to atmost 2 decimal places for easy comparisons
print(np.column_stack((Y_test,Y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```
